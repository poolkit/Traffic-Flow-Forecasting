```python
import json
import pandas as pd
```

```python
data = json.load(open('jobs_3217160_results_TrraficFlowAnalysis.json'))
```

```python
segment_results = []
for route in data['routes']:
    for segment in route['segmentResults']:
        try:
            segment_dict = {
                'segmentId': segment['segmentId'],
                'newSegmentId': segment['newSegmentId'],
                'speedLimit': segment['speedLimit'],
                'frc': segment['frc'],
                'streetName': segment['streetName'],
                'distance': segment['distance'],
                'shape': segment['shape']
            }
        except:
            pass
        segment_results.append(segment_dict)
        
df_segment_results = pd.DataFrame(segment_results)
```

```python
df_segment_results
```

```python
segment_time_results = []
for route in data['routes']:
    for segment in route['segmentResults']:
        for segment_time in segment['segmentTimeResults']:
            segment_time_dict = {
                'segmentId': segment['segmentId'],
                'timeSet': segment_time['timeSet'],
                'dateRange': segment_time['dateRange'],
                'harmonicAverageSpeed': segment_time['harmonicAverageSpeed'],
                'medianSpeed': segment_time['medianSpeed'],
                'averageSpeed': segment_time['averageSpeed'],
                'standardDeviationSpeed': segment_time['standardDeviationSpeed'],
                'travelTimeStandardDeviation': segment_time['travelTimeStandardDeviation'],
                'sampleSize': segment_time['sampleSize'],
                'averageTravelTime': segment_time['averageTravelTime'],
                'medianTravelTime': segment_time['medianTravelTime'],
                'travelTimeRatio': segment_time['travelTimeRatio'],
                'speedPercentiles': segment_time['speedPercentiles']
            }
            segment_time_results.append(segment_time_dict)
            
df_segment_time_results = pd.DataFrame(segment_time_results)
```
